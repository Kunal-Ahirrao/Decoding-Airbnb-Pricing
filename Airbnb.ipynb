```python
import numpy as np

# Define the IQR technique function
def iqr_technique(DFcolumn):
    # Remove NaN values before applying the IQR calculation
    DFcolumn_clean = DFcolumn.dropna()  # Drop NaN values
    
    # Check if the column is empty after removing NaN values
    if len(DFcolumn_clean) == 0:
        raise ValueError("The column is empty after dropping NaN values.")
    
    Q1 = np.percentile(DFcolumn_clean, 25)
    Q3 = np.percentile(DFcolumn_clean, 75)
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)  # Interquartile range
    return lower_range, upper_range

# Apply the IQR function to the 'price' column
lower_bound, upper_bound = iqr_technique(Airbnb_df['price'])

# Print the lower and upper bounds
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
```

```python
# Remove outliers from 'price' column based on the calculated lower and upper bounds
Airbnb_df = Airbnb_df[(Airbnb_df['price'] > lower_bound) & (Airbnb_df['price'] < upper_bound)]

# Check the number of rows before and after removing outliers
print(f"Original dataset size: {Airbnb_df.shape[0]}")
```

```python
#We outliers are removed from price column now check with boxplot and also check shape of new Dataframe!

sns.boxplot(x = Airbnb_df['price'])

print(Airbnb_df.shape)
```

# 5. How does availability (availability_365) affect pricing trends?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create bins for availability
Airbnb_df['availability_bins'] = pd.cut(
    Airbnb_df['availability_365'], 
    bins=[0, 50, 100, 150, 200, 250, 300, 365], 
    labels=['0-50', '51-100', '101-150', '151-200', '201-250', '251-300', '301-365']
)

# Calculate average price for each availability bin
avg_price_by_availability = Airbnb_df.groupby('availability_bins')['price'].mean()

plt.figure(figsize=(10, 6))
sns.lineplot(x=avg_price_by_availability.index, y=avg_price_by_availability.values, marker='o', color='red')
plt.xlabel('Availability (Days)')
plt.ylabel('Average Price ($)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Observations :
#### For Travelers:
More Expensive Prices for Medium Supply (101–150 Days):
The average prices become the highest for the listings with medium supply, that is, 101–150 days, meaning that this kind of listing may balance scarcity with demand.

 Duration Listings (301–365 Days):
Year-round listings tend to have the lowest average prices, likely due to increased competition among highly available properties.

 Price Affordable for Low Availability (0–50 Days):
Listings with very low availability, between 0 and 50 days, have lower average prices, potentially to stimulate bookings on the fewer number of dates available.

#### For Hosts:
Revenue Maximization with Moderate Availability:
Properties available for 101–150 days could result in higher prices, perhaps target towards high seasons or periods.

# 6. What are the characteristics of listings with the highest and lowest prices?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Define high and low price thresholds based on the original dataset
high_price_listings = Airbnb_df[Airbnb_df['price'] > Airbnb_df['price'].quantile(0.95)]
low_price_listings = Airbnb_df[Airbnb_df['price'] < Airbnb_df['price'].quantile(0.05)]

# Calculate raw counts for "Entire Home/Apt" in high and low price listings
high_price_entire_home_count = high_price_listings['room_type'].value_counts().get('Entire home/apt', 0)
low_price_entire_home_count = low_price_listings['room_type'].value_counts().get('Entire home/apt', 0)

# Calculate averages for other characteristics
summary_data = pd.DataFrame({
    'Room Type (Entire Home Count)': [
        high_price_entire_home_count,
        low_price_entire_home_count
    ],
    'Availability (Avg Days)': [
        high_price_listings['availability_365'].mean(),
        low_price_listings['availability_365'].mean()
    ],
    'Total Reviews (Avg)': [
        high_price_listings['total_reviews'].mean(),
        low_price_listings['total_reviews'].mean()
    ]
}, index=['High Price Listings', 'Low Price Listings'])

# Print the summary data
print("Summary Data for Heatmap:")
print(summary_data)

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(summary_data, annot=True, fmt=".2f", cmap="Reds", linewidths=0.5, cbar=True)
plt.xlabel('Characteristics')
plt.ylabel('Price Category')
plt.tight_layout()
plt.show()
```
